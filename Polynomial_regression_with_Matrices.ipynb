```python
#importing the important modules which we are going to use in
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from IPython.display import clear_output#to remove previous result from display
%matplotlib inline
```

```python
data=pd.read_csv('superconduct.csv')#read our csv file whose name is train.csv
```

```python
data=data.dropna(axis=0)#we just remove the rows which have nan or empty values
```

```python
data.describe()
```

```python
target=data['critical_temp']
features=data.drop('critical_temp',axis=1).copy()
#here we are going to add 1 more line for maximum value normalisation
features=features/features.max()
features=np.array(features)
target=target/target.max()
```

```python
#here we are copying our features in x_temp and doing x.T[0]^1 + x.T[1]^2+..+x.T[80]^81
x_temp=features.copy()
for i in range(len(x_temp.T)):
    x_temp.T[i]=x_temp.T[i]**(i+1)
    
```

```python
#here we reshape our target variable so that it is easily to perform various matrix operations
target=np.array(target)
print(target.shape)
target=target.reshape(1,len(target))
print(target.shape)
```

```python
x_temp.shape
```

```python

def curve(m,x,c):#eqation of curve y=M0X0^1+M1X1^2+...+MnXn^n+1+c
    return np.dot(m,x.T)+c
def error(m,x,c,y):#now we are going to make error function ==>mean squared error
    return np.mean((curve(m,x,c)-y)**2)
def derivative_slope(m,x,c,y):#now we are going to make derivative funtion for slope
    return 2*np.mean(np.multiply((curve(m,x,c)-y),x.T),axis=1)
def derivative_inter(m,x,c,y):#now we are going to make derivative funtion for slope
    return 2*np.mean(curve(m,x,c)-y)
def accuracy(error,y):
    return 100- (error/np.mean(y**2))*100
```

```python
#now lets check out our good work
m=np.random.randn(1,len(x_temp.T))#randomly initialize value of m with the help of normal distribution
c=random.random()#take any random number as c
alpha=1#learning rate
cost=[]
for i in range(0,5000):
    
    cost.append(error(m,x_temp,c,target))#append current error in cost, just to plot and visualize it
    m=m-alpha*(derivative_slope(m,x_temp,c,target)).T#m=m-learning_rate*d(error)/d(m)
    c=c-alpha*derivative_inter(m,x_temp,c,target)#c=c-learning_rate*d(error)/d(c)
    clear_output(True)#just to remove the previous printed output from the display
    print(i,'  Current error-->',error(m,x_temp,c,target),"  ",'Current Accuracy--->', accuracy(error(m,x_temp,c,target),target))
```

```python
plt.plot(cost)
plt.xlabel("number of iterations")
plt.ylabel("error")
plt.title("alpha=1")
```
